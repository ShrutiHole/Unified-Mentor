```python
data["Order Date"]=pd.to_datetime(data["Order Date"])
```

```python
data['Year']=data['Order Date'].dt.year
data['Month']=data['Order Date'].dt.month
```

```python
data["Total Revenue"]=data["Units Sold"]*data["Unit Price"]
```

```python
monthlytrend=data.groupby(["Month"]).agg({"Total Revenue":"sum","Units Sold":"sum","Order ID":"count"}).reset_index()
print("monthly_trend:\n",monthlytrend)
```

```python
yearlytrend=data.groupby(["Year"]).agg({"Total Revenue":"sum","Units Sold":"sum","Order ID":"count"}).reset_index()
print("yearly_trend:\n",yearlytrend)
```

```python
monthlyyearlytrend=data.groupby(["Year"]).agg({"Total Revenue":"sum","Units Sold":"sum","Order ID":"count"}).reset_index()
print("monthly_yearly_trend:\n",monthlyyearlytrend)
```

```python
correlationmatrix=data[["Total Revenue","Units Sold","Unit Price"]].corr()
print("correlation_matrix:\n",correlationmatrix)
```

```python
regionalsales=data.groupby(["Region"]).agg({"Total Revenue":"sum","Units Sold":"sum"}).reset_index()
print("Regional Sales Performance:\n",regionalsales)
```

```python
sns.scatterplot(x='Units Sold', y='Total Revenue', data=data)
plt.title('Total Revenue vs Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Total Revenue')
plt.legend(bbox_to_anchor=(1,1),loc=2)
```

```python
sns.scatterplot(x='Unit Price', y='Total Revenue', data=data)
plt.title('Total Revenue vs Unit Price')
plt.xlabel('Unit Price')
plt.ylabel('Total Revenue')
plt.legend(bbox_to_anchor=(1,1),loc=2)
```

```python
sns.scatterplot(x='Unit Price', y='Units Sold', data=data)
plt.title('Units Sold vs Unit Price')
plt.xlabel('Unit Price')
plt.ylabel('Units Sold')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data['Total Revenue'], bins=30, kde=True)
plt.title('Distribution of Total Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data['Units Sold'], bins=30, kde=True)
plt.title('Distribution of Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data['Unit Price'], bins=30, kde=True)
plt.title('Distribution of Unit Price')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(data=monthlytrend, x='Month', y='Total Revenue')
plt.title("Monthly Sales Trend")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(data=yearlytrend, x='Year', y='Total Revenue')
plt.title("Yearly Sales Trend")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=monthlytrend, x='Month', y='Total Revenue')
plt.title('Monthly Sales Trend')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=yearlytrend, x='Year', y='Total Revenue')
plt.title('Yearly Sales Trend')
plt.show()
```

```python
monthlysales = data.groupby(['Year', 'Month']).agg({"Total Revenue": 'sum',"Units Sold": 'sum'}).reset_index()
plt.figure(figsize=(12, 6))
plt.plot(pd.to_datetime(monthlysales[['Year', 'Month']].assign(Day=1)), monthlysales['Total Revenue'])
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
